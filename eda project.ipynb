2.Numerical-Categorical- create visualisations that compare the distribution of the numerical data across different categories of the categorical data.
a. You can plot graphs like barplot,boxplot,kdeplot violinplot even scatterplot
3. Categorical-Categorical
a.you can create cross-tabulations or contingency tables that show the distributions of values in one categorical column,grouped by the values in the other categorical column.

```python
#relationship between rating and company_manufacturer
df[df['company_manufacturer']== 'Soma']['rating'].plot(kind='kde',label='Soma')
df[df['company_manufacturer']== 'Fresco']['rating'].plot(kind='kde',label='Fresco')
df[df['company_manufacturer']== 'Arete']['rating'].plot(kind='kde',label='Arete')
df[df['company_manufacturer']== 'Bonnat']['rating'].plot(kind='kde',label='Bonnat')
df[df['company_manufacturer']== 'A. Morin']['rating'].plot(kind='kde',label='A.Morin')
plt.legend()
plt.show()

```

# Conclusions
Soma has more variability in its rating as compared to Arete. Arete variability is less as it is narrower and taller.Bonnat is slightly skewed to right.

```python
#relationship between rating and company_location
df[df['company_location']== 'U.S.A.']['rating'].plot(kind='kde',label='U.S.A')
df[df['company_location']== 'Canada']['rating'].plot(kind='kde',label='Canada')
df[df['company_location']== 'France']['rating'].plot(kind='kde',label='France')
df[df['company_location']== 'U.K.']['rating'].plot(kind='kde',label='U.K.')
df[df['company_location']== 'Italy']['rating'].plot(kind='kde',label='Italy')
plt.legend()
plt.show()
```

# Conclusions
ratings of the chocolate made in u.s.a. varies up and down.ratings of chocolate made in u.k. are slightly skewed to left.
